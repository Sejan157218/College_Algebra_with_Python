```python
# coefficient values
import math

a = 1
b = 5
c = 6

# vertex
vx = -b/(2*a)
vy = a*(vx**2) + b*vx + c
print("Vertex: (",vx,",",vy,")")

# Roots
d = b**2 - 4*a*c
if d>=0:
  root_1 = (-b + math.sqrt(d))/2*a
  root_2 = (-b - math.sqrt(d))/2*a
  print("Roots: x = ", root_1, " and x = ", root_2)
else:
  print("No Real Roots")
```

Graph a quadratic function

```python
import matplotlib.pyplot as plt
import numpy as np
import math

a = 1
b = 5
c = 6

# vertex
vx = -b/(2*a)
vy = a*(vx**2) + b*vx + c

xmin = -10
xmax = 10
ymin = -10
ymax = 10

points = 10*(xmax-xmin)
x = np.linspace(xmin, xmax, points)

fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin, xmax], [0,0], 'b')
plt.plot([0,0], [ymin, ymax], 'b')

y1 = a*x**2 + b*x +c
plt.plot(x, y1)

plt.plot([vx], [vy], 'ro')

d = b**2 -4*a*c
if d>=0:
  root_1 = (-b + math.sqrt(d))/2*a
  root_2 = (-b - math.sqrt(d))/2*a
  plt.plot([root_1, root_2],[0,0], 'go')
plt.show()
```

Sliders to show how a, b, and c affect the graph

```python

%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
import math


def f(a,b,c):
    # vertex
    vx = -b/(2*a)
    vy = a*(vx**2) + b*vx + c

    xmin = -10
    xmax = 10
    ymin = -10
    ymax = 10

    points = 10*(xmax-xmin)
    x = np.linspace(xmin, xmax, points)

    fig, ax = plt.subplots()
    plt.axis([xmin, xmax, ymin, ymax])
    plt.plot([xmin, xmax], [0,0], 'b')
    plt.plot([0,0], [ymin, ymax], 'b')

    y1 = a*x**2 + b*x +c
    plt.plot(x, y1)

    plt.plot([vx], [vy], 'ro')

    d = b**2 -4*a*c
    if d>=0:
      root_1 = (-b + math.sqrt(d))/2*a
      root_2 = (-b - math.sqrt(d))/2*a
      plt.plot([root_1, root_2],[0,0], 'go')
    # Set the equation as the title
    sa = str(a)
    sb = str(b)
    sc = str(c)
    h1 = "y = ", sa, "x**2 + ", sb, "x + ", sc
    h2 = ""
    for w in h1:
        h2 = h2 + w

    print(h2)
    # plt.title(str(h2))
    plt.show()


interactive_re = interactive(f, a=(1,9), b=(-9,9), c=(-9,9))
interactive_re
```
