```python
import matplotlib.pyplot as plt


old_demand_x = 1
new_demand_x = 14
old_price_y = 12
new_price_y = 3
old_supply_x = 11
new_supply_x = 1


xmin = 0
xmax = 15
ymin = 0
ymax = 15

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

plt.plot([old_demand_x,new_demand_x],[old_price_y,new_price_y], 'r')
plt.plot([old_supply_x,new_supply_x],[old_price_y,new_price_y], 'g')

plt.show()
```

# Caculate the price elasticity of Demand

percent change in demand / percent change in price

```python
old_demand_x = 5
new_demand_x = 6
old_price_y = 11
new_price_y = 10

demand_change = (new_demand_x - old_demand_x) / old_demand_x

price_change = (new_price_y - old_price_y) / old_price_y

e_number = demand_change / price_change

print(e_number)

if e_number>1:
  print("Demand id elastic")
elif e_number==1:
  print("demand is unitary or propotional")
elif e_number<1:
  print("Demand id Inelastic")
else:
  print("?")
```
