```python
from sympy import *

x,y = symbols('x y')

first = 2*x + y -1
second = x - 2*y + 7

# print(linsolve([first, second], (x,y)))

# parse finite set answer as coordinate pair
solution = linsolve([first, second], (x,y))
x_solution = solution.args[0][0]
y_solution = solution.args[0][1]
print("x_solution", x_solution, "y_solution", y_solution)
```

```python
from sympy import *

x,y = symbols('x y')

first = -x**2 - y + 10
second = 2*x**2 - 2*y - 4

# Solve - can be linear or nonlinear equations

solution = nonlinsolve([first, second], (x, y))

for i in range(len(solution.args)):
  x_solution = solution.args[i][0]
  y_solution = solution.args[i][1]

  # print("x_solution", x_solution, "y_solution", y_solution)

# Sympy syntax for equation equal to zero, ready to factor
y_first = Eq(first,0)
y1 = solve(y_first, y)


y_second = Eq(second, y)
y2 = solve(y_second,y)

x = symbols('x')
x_min = -10
x_max = 10
plot(y1[0], y2[0], (x, x_min, x_max))
```
